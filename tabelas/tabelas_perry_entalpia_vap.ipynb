```python
import tabula
import pandas as pd
```

```python
pdf_path = "perry.pdf"

dfs = tabula.read_pdf(pdf_path, stream=True, pages="155-162")
# read_pdf returns list of DataFrames
print(len(dfs))
```

```python
dfs[0].head()
```

```python
dfs0 = dfs[0].drop(["Unnamed: 2"], axis=1)
dfs0.head()
```

```python
dfs0 = dfs0.drop([0], axis=0)
dfs0.head()
```

```python
dfs0.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4', 'Tmin', 'DHmin', 'Tmax', 'DHmax']
dfs0.head()
```

```python
dfs0['C1'] = dfs0['C1'].astype(float).mul(1E07)
dfs0['DHmin'] = dfs0['DHmin'].astype(float).mul(1E07)
```
