```python
import tabula
import pandas as pd
```

```python
pdf_path = "perry.pdf"

dfs = tabula.read_pdf(pdf_path, stream=True, pages="128-134")
# read_pdf returns list of DataFrames
print(len(dfs))
```

```python
dfs[0]
```

```python
dfs0 = dfs[0].drop(["Unnamed: 0","Unnamed: 2", "Unnamed: 9", "Unnamed: 11", "Unnamed: 12", "Unnamed: 13"], axis=1).drop(0, axis=0)

dfs0.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4',
       'Tmin', 'Dmin', 'Tmax', 'Dmax']
```

```python
dfs0
```

```python
dfs[1]
```

```python
dfs1 = dfs[1].drop(["Unnamed: 0", "Unnamed: 9", "Unnamed: 10", "Unnamed: 11"], axis=1).drop(0, axis=0)
dfs1
```

```python
dfs1.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4',
       'Tmin', 'Dmin', 'Tmax', 'Dmax']
```

```python
dfs1
```

```python
dfs[2] = dfs[2].drop(['105'], axis=1)
dfs2_linha = dfs[2].columns
dfs2_linha_df = pd.DataFrame(dfs2_linha)
dfs2 = dfs2_linha_df.T
dfs2.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4',
       'Tmin', 'Dmin', 'Tmax', 'Dmax']
column_names = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4',
       'Tmin', 'Dmin', 'Tmax', 'Dmax']

dfs[2].columns=column_names
dfs[2]
merged_ds2 = pd.concat([dfs2, dfs[2]], ignore_index=True)
merged_ds2
```

```python
merged_ds2['compound'][0] = '105'
```

```python
merged_ds2
```

```python
dfs3 = dfs[3].drop(["Unnamed: 0", "Unnamed: 2", "Unnamed: 9", "Unnamed: 11", "Unnamed: 12", "Unnamed: 13"], axis=1).drop(0, axis=0)
dfs3.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4',
       'Tmin', 'Dmin', 'Tmax', 'Dmax']
dfs3
```

```python
dfs[4] = dfs[4].drop(['105'], axis=1)
dfs4_linha = dfs[4].columns
dfs4_linha_df = pd.DataFrame(dfs4_linha)
dfs4 = dfs4_linha_df.T
dfs4.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4',
       'Tmin', 'Dmin', 'Tmax', 'Dmax']
column_names = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4',
       'Tmin', 'Dmin', 'Tmax', 'Dmax']

dfs[4].columns=column_names
dfs[4]
merged_ds4 = pd.concat([dfs4, dfs[4]], ignore_index=True)
merged_ds4
```

```python
dfs5 = dfs[5].drop(["Unnamed: 0","Unnamed: 2", "Unnamed: 9", "Unnamed: 11", "Unnamed: 12", "Unnamed: 13"], axis=1).drop(0, axis=0)

dfs5.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4',
       'Tmin', 'Dmin', 'Tmax', 'Dmax']
dfs5
```

```python
dfs[6]
```

```python
dfs[6] = dfs[6].drop(['105','Unnamed: 0', 'Unnamed: 2', 'Unnamed: 1'], axis=1).drop([30], axis=0).drop(4,axis = 0)
```

```python

dfs6_linha = dfs[6].columns
dfs6_linha_df = pd.DataFrame(dfs6_linha)
dfs6 = dfs6_linha_df.T
dfs6.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4',
       'Tmin', 'Dmin', 'Tmax', 'Dmax']
column_names = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'C1', 'C2', 'C3', 'C4',
       'Tmin', 'Dmin', 'Tmax', 'Dmax']

dfs[6].columns=column_names
dfs[6]
merged_ds6 = pd.concat([dfs6, dfs[6]], ignore_index=True)
merged_ds6
```

```python
merged_ds6.loc[29, 'C1'] = "-13.851"
merged_ds6.loc[29, 'C3'] = "-0.0019124"
merged_ds6.loc[5, 'C2'] = "-0.003474"
merged_ds6.loc[5, 'C3'] = "0"
merged_ds6.loc[5, 'C4'] = "0"
```

```python
merged_ds6
```

```python
densidade_molar = pd.concat([dfs0, dfs1, merged_ds2,dfs3, merged_ds4, dfs5, merged_ds6], ignore_index=True)
densidade_molar
```

```python
for i in range(len(densidade_molar)):
    densidade_molar['MolWeight'][i] = float(densidade_molar['MolWeight'][i])
    densidade_molar['C1'][i] = float(densidade_molar['C1'][i])
    densidade_molar['C2'][i] = float(densidade_molar['C2'][i])
    densidade_molar['C3'][i] = float(densidade_molar['C3'][i])
    densidade_molar['C4'][i] = float(densidade_molar['C4'][i])
    densidade_molar['Tmin'][i] = float(densidade_molar['Tmin'][i])
    densidade_molar['Dmin'][i] = float(densidade_molar['Dmin'][i])
    densidade_molar['Tmax'][i] = float(densidade_molar['Tmax'][i])
    densidade_molar['Dmax'][i] = float(densidade_molar['Dmax'][i])
```

```python
densidade_molar.to_csv('densidade_molar.csv', index=False)
```
