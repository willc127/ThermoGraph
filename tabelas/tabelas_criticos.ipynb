```python
import tabula
import pandas as pd
```

```python
pdf_path = "perry.pdf"

dfs = tabula.read_pdf(pdf_path, stream=True, pages="218-224")
# read_pdf returns list of DataFrames
print(len(dfs))
```

```python
dfs[0].head()
```

```python
dfs0 = dfs[0].dropna(axis=1)
dfs0.head()
```

```python
dfs0.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'Tc', 'Pc', 'Vc', 'Zc', 'Acentric factor']
dfs0.head()
```

```python
dfs[1].head()
```

```python
dfs1 = dfs[1].dropna(axis=1)
dfs1.head()
```

```python
dfs1.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'Tc', 'Pc', 'Vc', 'Zc', 'Acentric factor']
dfs1.tail()
```

```python
dfs[2].head()
```

```python
dfs2_linha = dfs[2].columns
dfs2_linha_df = pd.DataFrame(dfs2_linha)
dfs2 = dfs2_linha_df.T
dfs2.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'Tc', 'Pc', 'Vc', 'Zc', 'Acentric factor']
column_names = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'Tc', 'Pc', 'Vc', 'Zc', 'Acentric factor']

dfs[2].columns=column_names
dfs[2]
merged_ds2 = pd.concat([dfs2, dfs[2]], ignore_index=True)
merged_ds2.tail()
```

```python
dfs3.columns = ['compound', 'Name', 'Formula', 'CAS','MolWeight', 'Tc', 'Pc', 'Vc', 'Zc', 'Acentric factor']
dfs3.head()
```
